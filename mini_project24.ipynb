```python
import os
import shutil
import cv2
from bing_image_downloader import downloader  # Make sure this package is installed

# Step 1: Create the input_images folder
input_folder = 'input_images'
os.makedirs(input_folder, exist_ok=True)

# Step 2: Download 100 images into input_images/search_keyword/
search_keyword = "hills"  # Change this to any search term
downloader.download(search_keyword, limit=100, output_dir=input_folder, adult_filter_off=True)

# Step 3: Move all images from subfolder to input_images/
subfolder_path = os.path.join(input_folder, search_keyword)

for file in os.listdir(subfolder_path):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        shutil.move(os.path.join(subfolder_path, file), os.path.join(input_folder, file))

# Remove the empty keyword-named folder
os.rmdir(subfolder_path)

print(f"✅ Downloaded and moved 100 '{search_keyword}' images to '{input_folder}'")

# Step 4: Resize all images to 50% and save to 'resized_images'
output_folder = 'resized_images'
os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(input_folder, file)
        img = cv2.imread(img_path)

        if img is None:
            print(f"⚠️ Skipped unreadable image: {file}")
            continue

        # Resize to 50%
        height, width = img.shape[:2] #get image height and width
        resized = cv2.resize(img, (width // 2, height // 2))

        # Save resized image
        cv2.imwrite(os.path.join(output_folder, file), resized)

print(f"✅ All images resized to 50% and saved in '{output_folder}'")
```
